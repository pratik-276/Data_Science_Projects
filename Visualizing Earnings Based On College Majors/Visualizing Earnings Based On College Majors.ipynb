## Question : What percent of majors are predominantly male? Predominantly female?

```python
total_men = recent_grads["Men"].sum()
total_men
```

```python
total_women = recent_grads["Women"].sum()
total_women
```

```python
total_students = total_men + total_women
total_students
```

```python
men_percentage = (100*total_men)/total_students
men_percentage
```

```python
women_percentage = (100*total_women)/total_students
women_percentage
```

So we can see that 57.52% are females and 42.48% are men.
